$
\LARGE\textbf{¿Qué es?}
$

<div class=text-justify>
El efecto fingering o interdigitación viscosa es el fenómeno por el cual $\textbf{un fluido se esparce en otro de forma ramificada}$, en general este comportamiento se obtiene a partir de inestabilidades, la $\textbf{anisotropía}$ es condición necesaria para que se produzca el efecto. Para fluidos newtonianos con diferencia significativa de densidad, el comportamiento se explica mediante la inestabilidad de Rayleigh-Taylor y para fluidos no newtonianos con $\textbf{diferencia significativa de viscosidad}$ mediante la inestabilidad de $\textbf{Saffman-Taylor}$. Utilizaremos los modelos matemáticos que describen $\textbf{Saffman-Taylor}$ para realizar un proyecto educativo que ilustre como la ramificación dendrítica varía en velocidad, tamaño de las ramas y de las subramas según parámetros del medio y de los fluidos involucrados, por lo tanto, a continuación se expondrá el modelo matemático antes mencionado.
</div>

$
\LARGE\textbf{¿Para qué sirve?}
$

<div class=text-justify>
El estudio de este fenómeno y sus causas sirven para expandir nuestro entendimiento de los fluidos: es sabido que no todos los fluidos se rigen por las mismas condiciones y propiedades, por lo que resulta necesario extender la teoría para poder asimilar singularidades que hasta el momento solo se conocen superficialmente. La aplicacion de esta teoría resulta clave para la optimización en actividades que involucren fluidos distintos, causando el $\textbf{patrón visual}$ que es característico del efecto. Entre estas actividades se encuentran: procesos de filtración, transporte de solutos, y la recuperación asistida de petróleo (ilustrada en la siguiente figura):

$
\LARGE\textbf{El entorno del efecto}
$

<div class=text-justify>
Antes de explicar las inestabilidades es fundamental comprender el ambiente en el que estas se producen, por lo que se introduce el concepto $\textbf{celda de Hele-Shaw}$. Estas celdas consisten en un montaje experimental comúnmente utilizado para producir y estudiar flujos potenciales bidimensionales y corresponden al espacio entre dos placas planas paralelas ligeramente separadas entre sí, tal como muestra la siguiente figura:
</div>

<img src="placas.png" width=300 height=60 />

Cuando un fluido se mueve entre las placas de la celda sigue la ley de Darcy:

$$
\nabla P = - \frac{\mu \vec{v}}{K} + \rho g
$$

Donde $\vec{v}$ es la velocidad del fluido, $\mu$ es la viscosidad del fluido, $\nabla P$ es el gradiente de presiones y  $K$ es un coeficiente específico del medio asociado a su capacidad de permitir el paso de un fluido. En el caso específico de la celda de Hele-Shaw tenemos $K=\frac{b^2}{12}$ donde b es la separación entre las placas. Reemplazando esto en la Ley de Darcy obtenemos nuestra ecuación guía:

$$
\nabla P = - \frac{12\mu \vec{v}}{b^2} + \rho g
$$

$
\LARGE\textbf{¿Cómo funciona?}
$

En el siguiente desarrollo se explicará en profundidad el funcionamiento de la inestabilidad de Saffman-Taylor.

Para comenzar es fundamental tener en cuenta que la inestabilidad de Saffman-Taylor es un fenómeno que ocurre cuando hya un desplazamiento $\textbf{horizontal}$ de un fluido que se expande sobre otro que tiene $\textbf{muy distinta viscosidad}$.

En primer lugar debemos notar, observando la Ley de Darcy, que el perfil de velocidades del fluido tendrá forma de parábola en el eje de la altura (separación entre las placas), dado que $\vec{v}$ depende de $b^2$, lo que gráficamente es de la forma:

<img src="velocidades_parabola.png" width=600 height=60 />

A continuación, hacemos un breve desarrollo matemático de la Ley de Darcy para el movimiento de un fluido sobre otro, expandiendo la divergencia:
$$
\begin{bmatrix} \frac{dp}{dx} \\ \frac{dp}{dy}  \\ \frac{dp}{dz}\end{bmatrix} = \begin{bmatrix}
  -\frac{(\mu_1-\mu_2)v_x}{K} + (\rho_1 - \rho_2) g x \\
   -\frac{(\mu_1-\mu_2)v_y}{K} + (\rho_1 - \rho_2)g y \\
    -\frac{(\mu_1-\mu_2)v_z}{K} + (\rho_1 - \rho_2)g z \\
\end{bmatrix}
$$

Si los dos fluidos tienen la misma densidad, la expresión matemática se reduce a:

$$
\begin{bmatrix} \frac{dp}{dx} \\ \frac{dp}{dy}  \\ \frac{dp}{dz}\end{bmatrix} = \begin{bmatrix}
  -\frac{(\mu_1-\mu_2)v_x}{K} \\
   -\frac{(\mu_1-\mu_2)v_y}{K}\\
    -\frac{(\mu_1-\mu_2)v_z}{K}\\
\end{bmatrix}  \rightarrow \nabla P = - \frac{12\mu \vec{v}}{b^2}
$$

A continuación, podemos aplicar gradiente a ambos lados obteniendo:

$$
\nabla^2 P = - \nabla \cdot \frac{12\mu \vec{v}}{b^2}
$$

Si estamos trabajando con fluidos $\textbf{incompresibles}$ se cumple que $\nabla \cdot \vec{v} = 0$. Por lo tanto, la expresión anterior quedaría de la forma:


$$
\nabla^2 P = 0
$$

Esta ecuación corresponde a la ecuación de Laplace, cuya solución es una *ecuación de onda*.

### Resolver la ecuación de Laplace requiere matemáticas muy complejas y entendimiento de las condiciones de borde, pero si tenemos la respuesta general, podemos encontrar muchas respuestas particulares cambiando los parámetros del problema

<img src="onda.png" width=600 height=60 />

<div class=text-justify>
En este caso los parámetros del problema corresponden a las viscosidades de los fluidos y la velocidad inicial. Siempre que los fluidos en cuestión no se puedan mezclar, también será un parámetro importante la tensión en la interfaz ($\gamma$), que corresponde a la energía necesaria para deformar una unidad de superficie.
    
Los parámetros de onda dependen de la frecuencia de las perturbaciones, si tenemos una frecuencia de perturbaciones constante, tendremos una longitud de onda constante, de no ser así, la solución de la ecuación de onda en base a los parámetros del problema nos indica que:    
</div>

$$
    0 \leq \lambda \leq 2\pi\sqrt{\frac{\gamma}{12(\mu_1-\mu_2)v_c} }
 $$

\LARGE\textbf{¿Cómo se ve?}
$

```python
%matplotlib inline
from IPython.display import HTML
from matplotlib import pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
```

```python
velocidad = 0.005
densidad = 2
viscosidad_fluido_izquierda = 1.5
viscosidad_fluido_derecha = 1
```

```python



# initializing a figure in
# which the graph will be plotted
fig = plt.figure()

# marking the x-axis and y-axis
axis = plt.axes(xlim=(-4, 4),
                ylim=(0, 4))

# initializing a line variable
line, = axis.plot([], [], lw=3)


# data which the line will
# contain (x, y)
def init():
    line.set_data([], [])
    return line,


def animate(i):
    global velocidad
    global densidad
    global viscosidad_fluido_izquierda
    global viscosidad_fluido_derecha
    
    
    y = np.linspace(0, 4, 1000)
    
    if viscosidad_fluido_izquierda == viscosidad_fluido_derecha:
        
        line.set_data(0, y)
        
        return line,
    
    lammda = np.sqrt(densidad / (12 * 9.18 * velocidad * (viscosidad_fluido_izquierda - viscosidad_fluido_derecha) ))
    print(lammda)
    
    escalador = np.linspace(4, 0.5, 1000)
    # plots a sine graph
    x =  escalador * (np.power(np.e, 1 + (velocidad * i) ) - np.e)* np.cos(y * lammda * (2 * np.pi))
    if lammda < 0:
        x = -x
    line.set_data(x, y)

    return line,


anim = FuncAnimation(fig, animate,
                     init_func=init,
                     frames=50,
                     interval=100,
                     blit=True)

HTML(anim.to_jshtml())
```
