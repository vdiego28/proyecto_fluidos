\Huge\textbf{Efecto Fingering}

$
\LARGE\textbf{Qué es}
$

<div class=text-justify>
El efecto fingering o interdigitación viscosa es el fenómeno en el cual $\textbf{un fluido se esparce en otro fluido de forma ramificada}$, en general este comportamiento se obtiene a partir de inestabilidades, la $\textbf{anisotropía}$ es condición necesaria para que se produzca el efecto. Para fluidos Newtonianos con diferencia significativa de densidad el comportamineto se explica mediante la inestabilidad de Rayleigh-Taylor y para fluidos no newtonianos con $\textbf{diferencia significativa de viscosidad}$ mediante la inestabilidad de $\textbf{Saffman-Taylor}$. Utilizaremos los modelos matemáticos que describen $\textbf{Saffman-Taylor}$ para realizar un proyecto educativo que ilustre como la ramificación dendrítica varía en velocidad, tamaño de las ramas y de las subramas según parámetros del medio y de los fluidos involucrados, por lo tanto, a continuación se expondrá el modelo matemático antes mencionado.
</div>

$
\LARGE\textbf{El entorno del efecto}
$

<div class=text-justify>
Antes de explicar las inestabilidades es fundamental comprender el ambiente en el que estas se producen, por lo que se introduce el concepto $\textbf{celda de Hele-Shaw}$. Estas celdas consisten en un montaje experimental comúnmente utilizado para producir y estudiar flujos potenciales bidimensionales y corresponden al espacio entre dos placas planas paralelas ligeramente separadas entre sí, tal como muestra la siguiente figura:
</div>

<img src="placas.png" width=300 height=60 />

Cuando un fluido se mueve entre las placas de la celda sigue la ley de Darcy:

$$
\nabla P = - \frac{\mu \vec{v}}{K} + \rho g
$$

Donde $\vec{v}$ es la velocidad del fluido, $\mu$ es la viscosidad del fluido, $\nabla p$ el gradiente de presiones y  $K$ es un coeficiente específico del medio asociado a su capacidad de permitir el paso de un fluido. En el caso específico de la celda de Hele-Shaw tenemos $K=\frac{b^2}{12}$ donde b es la separación entre las placas. Reemplazando esto en la Ley de Darcy obtenemos nuestra ecuación guía:

$$
\nabla P = - \frac{12\mu \vec{v}}{b^2} + \rho g
$$

$
\LARGE\textbf{Cómo funciona}
$

En el siguiente desarrollo se explicará en profundidad el funcionamiento de la inestabilidad de Saffman-Taylor.

Para comenzar es fundamental tener en cuenta que la inestabilidad de Saffman-Taylor es un fenómeno que ocurre cuando hya un desplazamiento $\textbf{horizontal}$ de un fluido que se expande sobre otro que tiene $\textbf{muy distinta viscosidad}$

En primer lugar debemos notar, observando la Ley de Darcy es que el perfil de velocidades del fluido tendrá forma de parábola en el eje de la altura (separación entre las placas), dado que $\vec{v}$ depende de $b^2$, lo que gráficamente es de la forma:

<img src="imagenes/velocidades_parabola.png" width=400 height=60 />

A continuación, hacemos un breve desarrollo matemático de la Ley de Darcy para el movimiento de un fluido sobre otro, expandiendo la divergencia:
$$
\begin{bmatrix} \frac{dp}{dx} \\ \frac{dp}{dy}  \\ \frac{dp}{dz}\end{bmatrix} = \begin{bmatrix}
  -\frac{(\mu_1-\mu_2)v_x}{K} + (\rho_1 - \rho_2) g x \\
   -\frac{(\mu_1-\mu_2)v_y}{K} + (\rho_1 - \rho_2)g y \\
    -\frac{(\mu_1-\mu_2)v_z}{K} + (\rho_1 - \rho_2)g z \\
\end{bmatrix}
$$

Si los dos fluidos tienen la misma densidad, la expresión matemática se reduce a:

$$
\begin{bmatrix} \frac{dp}{dx} \\ \frac{dp}{dy}  \\ \frac{dp}{dz}\end{bmatrix} = \begin{bmatrix}
  -\frac{(\mu_1-\mu_2)v_x}{K} \\
   -\frac{(\mu_1-\mu_2)v_y}{K}\\
    -\frac{(\mu_1-\mu_2)v_z}{K}\\
\end{bmatrix}  \rightarrow \nabla P = - \frac{12\mu \vec{v}}{b^2}
$$

A continuación, puedo aplicar gradiente a los dos lados obreniendo:

$$
\nabla^2 P = - \nabla \cdot \frac{12\mu \vec{v}}{b^2}
$$

Si estamos trabajando con fluidos $\textbf{incompresibles}$ se cumple que $\nabla \cdot \vec{v} = 0$, por lo tanto la expresión anterior quedaría de la forma:


$$
\nabla^2 P = 0
$$

Esta ecuación corresponde a la ecuación de Laplace, cuya solución es una *ecuación de onda*.

```python
import fluiddyn as fld
from fluidsim.solvers.ns2d.solver import Simul

params = Simul.create_default_params()

params.short_name_type_run = "test"

params.oper.nx = params.oper.ny = nh = 32
params.oper.Lx = params.oper.Ly = Lh = 10

delta_x = Lh / nh
params.nu_8 = 2.0 * params.forcing.forcing_rate ** (1.0 / 3) * delta_x ** 8

params.time_stepping.t_end = 10.0

params.init_fields.type = "dipole"

params.forcing.enable = True
params.forcing.type = "tcrandom"

params.output.sub_directory = "examples"

params.output.periods_print.print_stdout = 0.5

params.output.periods_save.phys_fields = 0.1 # dt_file
params.output.periods_save.spectra = 0.5
params.output.periods_save.spatial_means = 0.05
params.output.periods_save.spect_energy_budg = 0.5
params.output.periods_save.increments = 0.5

params.output.ONLINE_PLOT_OK = True

params.output.spectra.HAS_TO_PLOT_SAVED = True
params.output.spatial_means.HAS_TO_PLOT_SAVED = True
params.output.spect_energy_budg.HAS_TO_PLOT_SAVED = True
params.output.increments.HAS_TO_PLOT_SAVED = True

params.output.phys_fields.field_to_plot = "rot"


sim = Simul(params)

sim.time_stepping.start()
sim.output.phys_fields.animate(dt_frame_in_sec=0.1, dt_equations=0.1, repeat=False, save_file="hola.gif", tmax=10)

sim.output.phys_fields.plot()



fld.show()
```
